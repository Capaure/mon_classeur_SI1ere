## **Exercice 1 : Compléter la chaine de puissance avec effort et flux**
### Données
- Domaine électrique : \( U \) (Tension en volts), \( I \) (Courant en amperes).
- Domaine mécanique : \( F \) (Force en newtons), \( v \) (Vitesse en m/s).
- Formules utilisees :
  - Puissance électrique : $$ P = U \cdot I $$
  - Puissance mécanique (translation) : $$ P = F \cdot v $$

###  tableau

| Repère | Domaine physique | Variable d’effort       | Variable de flux         |
|--------|------------------|-------------------------|--------------------------|
| 1      | Électrique       | Tension (\( U \), V)    | Courant (\( I \), A)     |
| 2      | Électrique       | Tension (\( U \), V)    | Courant (\( I \), A)     |
| 3      | Mécanique        | Couple (\( C \), N·m)   | Vitesse angulaire (\( \omega \), rad/s) |
| 4      | Mécanique        | Force (\( F \), N)      | Vitesse linéaire (\( v \), m/s) |

## **Exercice 2 : Puissance de poussée utile**
### Données
- **Force de poussee** : *F* = 750 N
- **Vitesse** : *v* = 0,5 m/s
- **Formule utilisee** : $$ P = F \cdot v $$

### Calcul
$$
P_{\text{poussee}} = 750 \, \text{N} \times 0,5 \, \text{m/s} = 375 \, \text{W}
$$

Puissance totale avec deux helices :
$$
P_{\text{totale}} = 2 \cdot P_{\text{poussee}} = 2 \cdot 375 = 750 \, \text{W}
$$

## **Exercice 3 : Puissance reçue par les panneaux solaires**
### Données
- **Rendements** : 
  - *ηm* = 0,9
  - *ηₕ* = 0,62
  - *ηpv* = 0,22
  - *ηr* = 0,95
- Relation des rendement  :
$$
P_{\text{reçue}} = \frac{P_{\text{poussée}}}{\eta_{\text{M}} \cdot \eta_{\text{H}} \cdot \eta_{\text{R}} \cdot \eta_{\text{PV}}}
$$

### Calcul
$$
P_{\text{reçue}} = \frac{750}{0,9 \cdot 0,62 \cdot 0,95 \cdot 0,22} \approx 6014 \, \text{W}
$$

Puissance par panneau :
$$
P_{\text{panneau}} = \text{Irr} \cdot S = 780 \cdot 0,57 = 445,2 \, \text{W}
$$

Nombre de panneaux nécessaires :
$$
n = \frac{P_{\text{reçue}}}{P_{\text{panneau}}} = \frac{6014}{445,2} \approx 14 \, \text{panneaux}
$$

### Conclusion
il faut doinc 15 panneaux pour satisfaire l’exigence

## **Exercice 4 : Énergie embarquée dans les batteries**
### Données
- **Capacité** : *Q* = 115 Ah
- **Tension** : *U* = 12 V
- **Nombre de batteries** : 8
- **Formule utilisée** : $$ E = U \cdot Q $$

### Calcul
Énergie totale :
$$
E_{\text{totale}} = 8 \cdot (115 \cdot 12) = 11040 \, \text{Wh} = 11,04 \, \text{kWh}
$$

Énergie utilisable (80%) :
$$
E_{\text{utilisable}} = 0,8 \cdot 11,04 = 8,832 \, \text{kWh}
$$

## **Exercice 5 : Temps d’utilisation et distance parcourue**
### Données
- **Puissance consommée** : *P* = 1885 W
- **Vitesse** : *v* = 2,5 nœuds = 2,5 · 1,8 = 4,5 km/h


### Calcul
Temps d’utilisation :
$$
t = \frac{E_{\text{utilisable}}}{P} = \frac{8832}{1885} \approx 4,69 \, \text{heures}
$$

Distance parcourue :
$$
d = v \cdot t = 4,5 \cdot 4,69 \approx 21,1 \, \text{km}
$$
