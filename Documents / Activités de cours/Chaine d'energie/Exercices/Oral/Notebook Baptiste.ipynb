##  **_EXERCICE 1 :_** Bouilloire électrique

#### **1.**  À partir du principe de conservation de l'énergie, calculer l'énergie électrique absorbée par la bouilloire.

**Données :**  
- Puissance de la bouilloire : $P = 2 \, \text{kW} = 2000 \, \text{W}$
- Temps de fonctionnement : $t = 2 \, \text{min} \, 30 \, \text{s} = 150 \, \text{s}$
- Énergie thermique transférée à l'eau : $E_{\text{utile}} = 270 \, \text{kJ}$
- Énergie thermique perdue : $E_{\text{dégradée}} = 30 \, \text{kJ}$

Soit $"E_{\text{initiale}}"$ l'énergie électrique absorbée par la bouilloire, qui est égale à la somme des énergies transférées :  
$$
E_{\text{initiale}} = E_{\text{utile}} + E_{\text{dégradée}}
$$ 


**Calcul :**  
$$
E_{ \text {initiale}} = 270 \, \mathrm{kJ} + 30 \, \mathrm{kJ} = 300 \, \mathrm{kJ}
$$


Soit en formule python : 

```python
E_utile=270
E_dégradée=30
E_initiale=(E_utile)+(E_dégradée)
E_initiale
```

$E_{\text{initiale}}= 300\; \mathrm{kJ}$

------------------------------------

#### **2.** En déduire la formule permettant de calculer l'énergie électrique consommée par un appareil.

E = P \times t

- $t$ est le temps en secondes $\, \text{s}$.


**Calcul pour la bouilloire :**  
$$
E= 2000 \; \mathrm{W} \times 150\; \mathrm{s}
$$

Soit en formule python :

```python
Puissance=2000
Temps=150
Energie=(Puissance)*(Temps)
Energie
```

$E= 300000 \; \mathrm{J}$

**Calcul :**

##  **_EXERCICE 2 :_** Conversion d'énergie

Soit la conversion de $1\, \text {Wh}$ en Joules :
$$
P= \frac{E}{t}
$$


où :
- $E$ est l'énergie en Watt heure $\, \text {Wh}$
- $P$ est la puissance en watts $\, \text {W}$
- $t$ est le temps en heure $\, \text {h}$

**Par définition :**
$$
E= P\times t
$$

**Calcul :**
$$
1\; \mathrm{Wh}= 1\; \mathrm{W} \times 1\; \mathrm{h}=1\; \mathrm{W} \times 3600\; \mathrm{s}
$$

Soit en formule python :

```python
P=1
t=3600
E=(P)*(t)
E
```

$1\; \mathrm{Wh}= 3600\; \mathrm{J}$
